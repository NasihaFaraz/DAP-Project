```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\HP\Downloads\worldometer_data.csv")
data.head()
```

```python
sns.relplot(x="TotalCases",y="TotalDeaths",data=data)
```

```python
top_deaths = data.nlargest(10, 'TotalDeaths')[['Country/Region', 'TotalDeaths']]
countries = top_deaths['Country/Region']
total_deaths = top_deaths['TotalDeaths']

plt.figure(figsize=(12, 8))
plt.bar(countries, total_deaths, color='salmon', edgecolor='black')
plt.title('Top 10 Countries by Total Deaths', fontsize=16)
plt.xlabel('Country/Region', fontsize=14)
plt.ylabel('Total Deaths', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
 total_deaths = data['TotalDeaths']

plt.figure(figsize=(10, 6))
plt.hist(total_deaths, bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Distribution of Total Deaths', fontsize=16)
plt.xlabel('Total Deaths', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

top_countries = data.nlargest(10, 'Population')[['Country/Region', 'Population']]
countries = top_countries['Country/Region']
populations = top_countries['Population']

plt.pie(populations, labels=countries, autopct='%1.1f%%', startangle=140,colors=[ 'lightblue', 'lightgreen', 'lightcoral', 'lightpink','lightyellow', 'lightgoldenrodyellow', 'lightcyan', 'lavender','mistyrose', 'peachpuff'])
plt.title('Top 10 Countries by Population')
plt.axis('equal') 
plt.show()
```
